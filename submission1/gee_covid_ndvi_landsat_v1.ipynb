```python
from IPython.display import Image
import ee, folium
#import folium.colormap as cm
import branca
import branca.colormap as cm
ee.Initialize()
print (folium.__file__)
print (folium.__version__)
```

```python
collection20_b5 = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA').select('B5').filterDate('2020-03-23', '2020-05-31').mean()
collection20_b4 = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA').select('B4').filterDate('2020-03-23', '2020-05-31').mean()
ndvi20 = collection20_b5.subtract(collection20_b4).divide(collection20_b5.add(collection20_b4)).rename('NDVI')

collection19_b5 = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA').select('B5').filterDate('2019-03-23', '2019-05-31').mean()
collection19_b4 = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA').select('B4').filterDate('2019-03-23', '2019-05-31').mean()
ndvi19 = collection19_b5.subtract(collection19_b4).divide(collection19_b5.add(collection19_b4)).rename('NDVI')

no2_diff = ndvi19.subtract(ndvi20)
no2_diff = no2_diff.divide(ndvi19).multiply(100)


roi = ee.Geometry.Rectangle([60, 5, 98.5+5, 37])
band_viz = {
  'min': -100,
  'max': 100,
  #'palette': ['yellow', 'orange', 'red']
  'palette': ['blue', 'white', 'green']
}
colormap = cm.LinearColormap(['blue', 'white', 'green'], vmin=-100, vmax=100)
#colormap = cm.linear.YlOrRd_09.scale(0, 0.0002).to_step(10)
colormap.caption = '% change in NDVI'
# no2_diff = collection19.subtract(collection20)
# no2_diff = no2_diff.divide(collection19).multiply(100)
mapid = no2_diff.clip(roi).getMapId(band_viz)

map = folium.Map(location=[24.11,75.27+5],zoom_start=4, height=550,width=600)

folium.TileLayer(
    tiles=mapid['tile_fetcher'].url_format,
    attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    overlay=True,
    name='NDVI',
    opacity=0.45,
  ).add_to(map)

map.add_child(folium.LayerControl())
map.add_child(colormap)
#folium.map.LayerControl('topleft', collapsed=False).add_to(map)
title_html = '''
             <h3 align="center" style="font-size:20px"><b>NDVI % change (2019-2020) </b></h3>
             '''
map.get_root().html.add_child(folium.Element(title_html))

map.save('ndvi_india.html')
map
```
