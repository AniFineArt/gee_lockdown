```python
collection20 = ee.ImageCollection('ECMWF/CAMS/NRT').select('particulate_matter_d_less_than_25_um_surface').filterDate('2020-03-23', '2020-05-31').mean()
collection19 = ee.ImageCollection('ECMWF/CAMS/NRT').select('particulate_matter_d_less_than_25_um_surface').filterDate('2019-03-23', '2019-05-31').mean()
no2_diff = collection19.subtract(collection20)
no2_diff = no2_diff.divide(collection19).multiply(100)

roi = ee.Geometry.Rectangle([60, 5, 98.5+5, 37])
band_viz = {
  'min': -100,
  'max': 100,
  #'palette': ['yellow', 'orange', 'red']
  'palette': ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']
}
colormap = cm.LinearColormap(['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red'], vmin=-100, vmax=100)
#colormap = cm.linear.YlOrRd_09.scale(0, 0.0002).to_step(10)
colormap.caption = '% change in column number density'
no2_diff = collection19.subtract(collection20)
no2_diff = no2_diff.divide(collection19).multiply(100)
mapid = no2_diff.clip(roi).getMapId(band_viz)

map = folium.Map(location=[24.11+2,75.27+5],zoom_start=4, height=550,width=600)

folium.TileLayer(
    tiles=mapid['tile_fetcher'].url_format,
    attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    overlay=True,
    name='PM2.5',
    opacity=0.2,
  ).add_to(map)

map.add_child(folium.LayerControl())
map.add_child(colormap)
#folium.map.LayerControl('topleft', collapsed=False).add_to(map)
title_html = '''
             <h3 align="center" style="font-size:20px"><b>PM2.5 % change (2019-2020) </b></h3>
             '''
map.get_root().html.add_child(folium.Element(title_html))

map.save('pm2.5_india.html')

os.system("wkhtmltopdf --dpi 1000 pm2.5_india.html pm2.5_india.pdf")
os.system("pdfcrop --margins '0 0 0 0' pm2.5_india.pdf")
map
```
