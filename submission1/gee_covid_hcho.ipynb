```python
from IPython.display import Image
import ee, folium
#import folium.colormap as cm
import branca
import branca.colormap as cm
ee.Initialize()
print (folium.__file__)
print (folium.__version__)
```

```python
collection20 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_HCHO').select('tropospheric_HCHO_column_number_density').filterDate('2020-03-23', '2020-06-30').mean()
collection19 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_HCHO').select('tropospheric_HCHO_column_number_density').filterDate('2019-03-23', '2019-06-30').mean()
band_viz = {
  'min': 0.0,
  'max': 0.0003,
  #'palette': ['yellow', 'orange', 'red']
  'palette': ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']
}
roi = ee.Geometry.Rectangle([65, 5, 98.5, 37])
```

```python
colormap = cm.LinearColormap(['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red'], vmin=0.0, vmax=0.05)
#colormap = cm.linear.YlOrRd_09.scale(0, 0.0002).to_step(10)
colormap.caption = 'NO2'
no2_diff = collection20.subtract(collection19)
#no2_diff = no2_diff.divide(collection19)
mapid = collection20.clip(roi).getMapId(band_viz)

map = folium.Map(location=[24.11,75.27],zoom_start=4, height=550,width=600)

folium.TileLayer(
    tiles=mapid['tile_fetcher'].url_format,
    attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    overlay=True,
    name='CO',
    opacity=0.5,
  ).add_to(map)

map.add_child(folium.LayerControl())
map.add_child(colormap)
#folium.map.LayerControl('topleft', collapsed=False).add_to(map)
title_html = '''
             <h3 align="center" style="font-size:20px"><b>CO 2020</b></h3>
             '''
map.get_root().html.add_child(folium.Element(title_html))
map.save('plot_data.html')
map
```

```python
band_viz = {
  'min': -30,
  'max': 30,
  #'palette': ['yellow', 'orange', 'red']
  'palette': ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']
}
colormap = cm.LinearColormap(['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red'], vmin=-30, vmax=30)
#colormap = cm.linear.YlOrRd_09.scale(0, 0.0002).to_step(10)
colormap.caption = '% change in column number density'
no2_diff = collection19.subtract(collection20)
no2_diff = no2_diff.divide(collection19).multiply(100)
mapid = no2_diff.clip(roi).getMapId(band_viz)

map = folium.Map(location=[24.11,75.27],zoom_start=4, height=550,width=600)

folium.TileLayer(
    tiles=mapid['tile_fetcher'].url_format,
    attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    overlay=True,
    name='HCHO',
    opacity=0.5,
  ).add_to(map)

map.add_child(folium.LayerControl())
map.add_child(colormap)
#folium.map.LayerControl('topleft', collapsed=False).add_to(map)
title_html = '''
             <h3 align="center" style="font-size:20px"><b>HCHO % change (2019-2020) </b></h3>
             '''
map.get_root().html.add_child(folium.Element(title_html))

map.save('plot_data.html')
map
```
